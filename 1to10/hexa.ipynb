```python
import numpy as np
def on_a_scale_of_1_to_10(frac:"list of fractions", lw:"line width"=80) -> None:
    """Given a list of fractions between [0,1], visualize in ASCII.

    If given is 1D numpy array, apply line width.
    If given is 2D numpy array, print as is.
    If given is higher-D or not numpy array, error may occur.
    """
    if sum(frac.shape) == 0:
        return
    if len(frac.shape) == 2:
        m, n = frac.shape
        on_a_scale_of_1_to_10(frac.reshape(m*n), n)
    if len(frac.shape) == 1:
        row = frac[:lw]
        clipped = np.clip(row, 0, 1)
        rounded = np.round(clipped*9).astype('int')
        print(''.join(" .:-=+*#%@"[r] for r in rounded))
        on_a_scale_of_1_to_10(frac[lw:], lw)

A = np.linspace(0, 16, 70)[:, np.newaxis]
B = np.linspace(0, 8, 100)[np.newaxis, :]
sin1 = np.sin(A + B)
sin2 = np.sin(1.366*B - 0.366*A)
sin3 = np.sin(1.366*A - 0.366*B)
sin4 = (sin1 * sin2 * sin3)**2 * 2.5
on_a_scale_of_1_to_10(sin4)
```
